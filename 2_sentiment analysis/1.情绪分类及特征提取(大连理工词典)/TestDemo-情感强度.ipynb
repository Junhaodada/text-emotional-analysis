# 对情感强度进行特征提取

情感强度分为1,3,5,7,9五档，9表示强度最大，1为强度最小。

情感强度：Qd0~Qd9


```python
import pandas as pd
import jieba
import time


#获取数据集
f= open('庆余年220.csv',encoding='utf8')
weibo_df=pd.read_csv(f)
# print(weibo_df.head())
```

```python
df=pd.read_excel('大连理工大学中文情感词汇本体NAU.xlsx')
df = df[['词语', '词性种类', '词义数', '词义序号', '情感分类', '强度', '极性']]
df.head()
```

```python
Qd0=[]
Qd1=[]
Qd2=[]
Qd3=[]
Qd4=[]
Qd5=[]
Qd6=[]
Qd7=[]
Qd8=[]
Qd9=[]

for idx,row in df.iterrows():
    if row['强度'] in [0] :
        Qd0.append(row['词语'])
    if row['强度'] in [1]:
        Qd1.append(row['词语'])
    if row['强度'] in [2]:
        Qd2.append(row['词语'])
    if row['强度'] in [3]:
        Qd3.append(row['词语'])
    if row['强度'] in [4]:
        Qd4.append(row['词语'])
    if row['强度'] in [5]:
        Qd5.append(row['词语'])
    if row['强度'] in [6]:
        Qd6.append(row['词语'])
    if row['强度'] in [7]:
        Qd7.append(row['词语'])
    if row['强度'] in [8]:
        Qd8.append(row['词语'])
    if row['强度'] in [9]:
        Qd9.append(row['词语'])

print(len(Qd0)+len(Qd1)+len(Qd2)+len(Qd3)+len(Qd4)+len(Qd5)+len(Qd6)+len(Qd7)+len(Qd8)+len(Qd9))
```

```python
#添加自定义词典和停用词
#jieba.load_userdict("user_dict.txt")
stop_list = pd.read_csv('stop_words.txt',
                        engine='python',
                        encoding='utf-8',
                        delimiter="\n",
                        names=['t'])

#获取重命名t列的值
stop_list = stop_list['t'].tolist()

def txt_cut(juzi):
    return [w for w in jieba.lcut(juzi) if w not in stop_list]     #可增加len(w)>1
```

```python
def emotion_caculate(text):
    qd0=0
    qd1=0
    qd2=0
    qd3=0
    qd4=0
    qd5=0
    qd6=0
    qd7=0
    qd8=0
    qd9=0
    
    qd0_list = []
    qd1_list = []
    qd2_list = []
    qd3_list = []
    qd4_list = []
    qd5_list = []
    qd6_list = []
    qd7_list = []
    qd8_list = []
    qd9_list = []
    
    wordlist=txt_cut(text)
    wordset=set(wordlist)
    wordfreq=[]
#     print(wordset)
    for word in wordset:
        freq=wordlist.count(word)
#         print('word:',word,'freq:',freq)
        if word in Qd0:
            qd0+=freq
            qd0_list.append(word)
        if word in Qd1:
            qd1+=freq
            qd1_list.append(word)
        if word in Qd2:
            qd2+=freq
            qd2_list.append(word)
        if word in Qd3:
            qd3+=freq
            qd3_list.append(word)
        if word in Qd4:
            qd4+=freq
            qd4_list.append(word)
        if word in Qd5:
            qd5+=freq
            qd5_list.append(word)
        if word in Qd6:
            qd6+=freq
            qd6_list.append(word)
        if word in Qd7:
            qd7+=freq
            qd7_list.append(word)
        if word in Qd8:
            qd8+=freq
            qd8_list.append(word)
        if word in Qd9:
            qd9+=freq
            qd9_list.append(word)
        
    emotion_info={
        'length':len(wordlist),
        'qd0':qd0,
        'qd1':qd1,
        'qd2':qd2,
        'qd3':qd3,
        'qd4':qd4,
        'qd5':qd5,
        'qd6':qd6,
        'qd7':qd7,
        'qd8':qd8,
        'qd9':qd9
    }
    print(qd0_list,qd1_list,qd2_list,qd3_list,qd4_list,qd5_list,qd6_list,qd7_list,qd8_list,qd9_list) 
    indexs = ['length', 'qd0', 'qd1', 'qd2', 'qd3','qd4','qd5','qd6', 'qd7', 'qd8', 'qd9']
    #return pd.Series(emotion_info, index=indexs), anger_list, disgust_list, fear_list, sad_list, surprise_list, good_list, happy_list
    return pd.Series(emotion_info, index=indexs)
 
```

```python
start=time.time()
emotion_df=weibo_df['review'].apply(emotion_caculate)
end=time.time()
print(end-start)
print(emotion_df.head())
```

```python
output_df=pd.concat([weibo_df,emotion_df],axis=1)
print(output_df)
output_df.to_csv('庆余年220_qd.csv',encoding='utf_8_sig',index=False)
```
